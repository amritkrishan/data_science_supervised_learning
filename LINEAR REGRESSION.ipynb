```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=(20.0,10.0)

data=pd.read_csv('/Users/a0k01jq/Downloads/ML/headbrain.csv')
print(data.shape)
data.head()
```

```python
X=data['Head Size(cm^3)'].values
Y=data['Brain Weight(grams)'].values
```

```python
mean_x = np.mean(X)
mean_y = np.mean(Y)
n = len(X)
numer=0
denom=0
for i in range(n):
    numer+=(X[i]-mean_x)*(Y[i]-mean_y)
    denom+=(X[i]-mean_x)**2
# Values for m and c
b1=numer/denom
b0=mean_y - (b1*mean_x)
print(b1,b0)
```

```python
# Plotting values on graph
max_x=np.max(X)+100
min_x=np.min(X)-100

# Calc line values x and y
x=np.linspace(min_x, max_x, 1000)
y=b0+b1*x

plt.plot(x,y,color='#58b970', label='Regression Line')
plt.scatter(X,Y,color='#ef5423', label='Scatter Plot')

plt.xlabel('Head Size')
plt.ylabel('Brain Weight')

plt.legend()
plt.show()
```

```python
# How good the model is?
ss_t = 0
ss_r = 0
for i in range(n):
    y_pred=b0+b1*X[i]
    ss_t += (Y[i]-mean_y)**2
    ss_r += (Y[i]-y_pred)**2
r2 = 1 - (ss_r/ss_t)
print(r2)
```
